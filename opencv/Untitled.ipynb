```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('dashcam-road-rage.jpg')
```

```python
cv2.imshow('image',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
grey = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(grey,cmap='Greys')
```

```python
feats=cv2.goodFeaturesToTrack(grey,25,0.01,10)
```

```python
for p in feats.squeeze():
    cv2.circle(img,(p[0],p[1]),3,(255,0,0,0))
```
